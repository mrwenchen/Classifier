```python
def artist_feature(movement, gender):
    return {'movement': movement, 'gender': gender}
```

```python
zipped_feature_and_labels = zip(dataset.movement, list_of_genders, dataset.price)
```

```python
feature_sets = [ (artist_feature(movement.split(',')[0], gender), price) for movement, gender, price in zipped_feature_and_labels]
feature_sets
```

```python
import math
split_num = math.floor(len(feature_sets) * 0.8)
split_num
```

```python
training_set = feature_sets[:split_num]
testing_set = feature_sets[split_num:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
print(dataset.movement[133])
print(list_of_genders[133])
```

```python
classifier.classify(artist_feature('Punk', 'mostly_male'))
```

```python
print(dataset.price[133])
```

```python
# Jasmine has 20% accuracy for the classifier with the same dataset bacause she sorts gender into 3 categories rather 6 with
# addition of mostly_male, mostly_female and andy

print(nltk.classify.accuracy(classifier, testing_set))
```

```python
classifier.show_most_informative_features(20)
```
