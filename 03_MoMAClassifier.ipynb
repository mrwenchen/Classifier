```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
def setup_feature(title):
    if isinstance(title, (float, int)):
        title = str(title)
        keywords = [ x for x in title ]
        while len(keywords) < 3:
            keywords.append('none')
        return {'first_keyword': keywords[0], 'second_keyword': keywords[1], 'third_keyword': keywords[2]}
    else:
        title = title.lower().replace('(', '').replace(')', '')
        list_of_words = title.split(' ')
        keywords = [ x for x in list_of_words if x not in stopwords.words('english') and not x.isdigit()]
        while len(keywords) < 2:
            keywords.append('none')
        return {'first_keyword': keywords[0], 'second_keyword': keywords[1]}
```

```python
def setup_feature_2(origin):
    if origin is None:
        return 'N/A'
    elif isinstance(origin, (float, int)):
        return 'N/A'
    else:
        origin = origin.replace('(', '').replace(')', '')
        list_of_words = origin.split(', ') 
        # print(list_of_words)
        keywords = [ x for x in list_of_words ]
        return keywords[0]

print(setup_feature_2(''))
```

```python
zipped_feature_and_labels = zip(artworks.Title, artworks.ArtistBio)
```

```python
setup_feature(291)
```

```python
feature_sets = [ (setup_feature(title), setup_feature_2(origin)) for title, origin in zipped_feature_and_labels]
feature_sets
```

```python
import math

split_num = math.floor(len(feature_sets) * 0.8)
split_num
```

```python
training_set = feature_sets[:split_num]
testing_set = feature_sets[split_num:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
artworks.Title[2314]
```

```python
classifier.classify(setup_feature('Evaporating Dish'))
```

```python
print(nltk.classify.accuracy(classifier, testing_set))
```
