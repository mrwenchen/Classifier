```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

def setup_feature(title):
    title = title.lower()
    list_of_words = title.split()
    keywords = [ x for x in list_of_words if x not in stopwords.words('english') and not x.isdigit() and '&' not in list_of_words]
    while len(keywords) < 3:
        keywords.append('none')
    return {'first_keyword': keywords[0], 'second_keyword': keywords[1], 'third_keyword': keywords[2]}
```

```python
zipped_feature_and_labels = zip(series.title, series.listed_in)
```

```python
feature_sets = [ (setup_feature(title), category) for title, category in zipped_feature_and_labels]
feature_sets
```

```python
import math

split_num = math.floor(len(feature_sets) * 0.8)
split_num
```

```python
training_set = feature_sets[:split_num]
testing_set = feature_sets[split_num:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
series.title[230]
```

```python
classifier.classify(setup_feature('The River Runner'))
```

```python
series.listed_in[230]
```

```python
print(nltk.classify.accuracy(classifier, testing_set))
```

```python
classifier.show_most_informative_features(20)
```
